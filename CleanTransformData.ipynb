## 3. Missing Data

```python
# Count missing values per column
missing_values = df.isnull().sum().sort_values(ascending=False)

# Show percentage of missing values
missing_percentage = (missing_values / len(df)) * 100

# Combine both into a summary
missing_summary = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percentage.round(2)
})

display(missing_summary)
```

```python
cols_to_check = ['transactions', 'transactionRevenue', 'bounces', 'newVisits']

summary = []

for col in cols_to_check:
    unique_vals = df[col].dropna().unique()
    summary.append({
        'Column': col,
        'Num Unique Values': len(unique_vals),
        'Unique Values': sorted(unique_vals),
        'Contains 0?': 0 in unique_vals
    })

summary_df = pd.DataFrame(summary)
display(summary_df)
```

Missing values in transactions, bounces, and newVisits likely mean the event didn‚Äôt happen. Checked the unique values and no 0s ‚Äî only values like 1.0, 2.0, etc. So, NaNs just mean 0 in this case. Replace NaNs with zero

```python
df[['transactions', 'transactionRevenue', 'bounces', 'newVisits']] = df[['transactions', 'transactionRevenue', 'bounces', 'newVisits']].fillna(0)
```

\
**pageviews** only 10 rows missing. Since every session should have at least one pageview, columns filled with median.

```python
# Check the visits for those missing rows
print(df[df['pageviews'].isnull()]['visits'].value_counts())
```

```python
median_pageviews = df['pageviews'].median()
df['pageviews'] = df['pageviews'].fillna(median_pageviews)
```

**timeOnSite** is missing in a lot of rows. If a user bounced (bounces == 1), it's normal that time isn‚Äôt recorded

```python
# Fill timeOnSite with 0 where bounces == 1 and timeOnSite is missing
df.loc[(df['timeOnSite'].isnull()) & (df['bounces'] == 1), 'timeOnSite'] = 0

# Check how many are still missing (should be those where bounces == 0)
remaining = df['timeOnSite'].isnull().sum()
print(f"üïê Remaining missing timeOnSite values: {remaining}")
```

```python
# Show first 10 rows where timeOnSite is still NaN
df[df['timeOnSite'].isnull()].head(10)
```

Before we decide how to fill the remaining missing timeOnSite values, let's compare those rows with the rest of the dataset. If their engagement (like pageviews or transactions) is clearly lower, maybe we shouldn't use the median ‚Äî maybe the 25th percentile makes more sense. \
This ensures no overestimating session duration for users who likely weren‚Äôt very engaged. If their behavior suggests short, low-effort visits, filling in the average user‚Äôs session time (the median) might give an unrealistically high signal. Using the lower quartile (25th percentile) would better reflect who they likely were.

```python
# Subset of rows where timeOnSite is still missing
missing_time = df[df['timeOnSite'].isnull()]

# Compare describe() of full vs missing subset
print("Full Dataset")
display(df[['transactions', 'pageviews', 'bounces']].describe())

print("Subset with missing timeOnSite")
display(missing_time[['transactions', 'pageviews', 'bounces']].describe())

# Also look at 25th percentile
q25_time = df['timeOnSite'].quantile(0.25)
print(f"25th percentile of timeOnSite: {q25_time}")

# Median for reference
median_time = df['timeOnSite'].median()
print(f"Median of timeOnSite: {median_time}")
```

The 125 rows with missing timeOnSite clearly show lower engagement:

Pageviews: 1.61 vs. 5.80 (full)

Transactions: 0.008 vs. 0.031 (full)

Bounces: All 0 (i.e., no bounce occurred)

This confirms these were low-effort sessions, likely users who clicked around briefly or dropped quickly.

Since these users were minimally engaged and their stats are much lower than average, we filled missing timeOnSite values with 0.0 (25th percentile). This avoids inflating their activity.

```python
# Fill remaining nulls in timeOnSite with 0.0 based on low engagement
df['timeOnSite'] = df['timeOnSite'].fillna(0.0)
```

## 7. Unique Identifier Key

```python
# Count how many fullVisitorIds are duplicated
duplicate_visitors = df['fullVisitorId'].duplicated().sum()
print(f"Duplicate fullVisitorIds: {duplicate_visitors}")

# Count how many visitIds are duplicated
duplicate_visits = df['visitId'].duplicated().sum()
print(f"Duplicate visitIds: {duplicate_visits}")
```

```python
# Count unique visitor+visit combinations
unique_combos = df[['fullVisitorId', 'visitId']].drop_duplicates().shape[0]
print(f"Unique (fullVisitorId, visitId) pairs: {unique_combos}")

# Count duplicate combinations
duplicate_combos = df.duplicated(subset=['fullVisitorId', 'visitId']).sum()
print(f"Duplicate (fullVisitorId, visitId) pairs: {duplicate_combos}")
```

```python
# Get the 77 duplicate (visitor, session) pairs
dupes = df[df.duplicated(subset=['fullVisitorId', 'visitId'], keep=False)]

# Group and check differences
grouped = dupes.groupby(['fullVisitorId', 'visitId']).agg('nunique')

# Show a few examples where there's variation
grouped[grouped.max(axis=1) > 1]
```

While investigating duplicate visitId + fullVisitorId pairs, I noticed that many of them have identical values for most fields, except for time-based columns like visitStartTime, timeOnSite, and pageviews. This suggests that they are actually the same user session, but were split into two rows in the dataset.

This happens because many web analytics tools (like Google Analytics) reset sessions at midnight. If a user is active on the website across midnight, their activity can end up being split into two parts ‚Äî even if the session ID (visitId) stays the same.

Also, since the dataset timestamps are stored in UTC, and our users are in the US (Pacific Time), midnight in UTC could be early evening in US time ‚Äî making these split sessions even more likely.

```python
# Get all duplicated session pairs (same fullVisitorId + visitId)
dupes = df[df.duplicated(subset=['fullVisitorId', 'visitId'], keep=False)].copy()

# Convert visitStartTime to UTC datetime
dupes['visitStartTime_utc'] = pd.to_datetime(dupes['visitStartTime'], unit='s')

# Convert UTC to US Pacific time (automatically handles PDT/PST)
dupes['visitStartTime_us'] = dupes['visitStartTime_utc'].dt.tz_localize('UTC').dt.tz_convert('US/Pacific')

# Sort the duplicates so matching sessions appear together
dupes_sorted = dupes.sort_values(by=['fullVisitorId', 'visitId', 'visitStartTime'])

# View the first few rows
dupes_sorted[['fullVisitorId', 'visitId', 'visitStartTime', 'visitStartTime_utc', 'visitStartTime_us', 'pageviews', 'timeOnSite', 'transactions']].head(20)
```

After converting the timestamps to US time, we saw that these rows often represent activity that happened just before and after midnight.

This explains why they were stored in two parts ‚Äî even though they belong to the same session. To avoid double-counting, we grouped them by fullVisitorId and visitId and combined each pair into one row, summing values like pageviews, timeOnSite, and transactions

```python
# Define how to aggregate each column
agg_rules = {
    'visitNumber': 'first',
    'visitStartTime': 'min',  # keep the earliest timestamp
    'date': 'min',
    'channelGrouping': 'first',
    'deviceCategory': 'first',
    'continent': 'first',
    'country': 'first',
    'source': 'first',
    'medium': 'first',
    'visits': 'sum',
    'pageviews': 'sum',
    'timeOnSite': 'sum',
    'bounces': 'sum',
    'transactions': 'sum',
    'transactionRevenue': 'sum',
    'newVisits': 'first'
}

# Group by visitor ID and visit ID to merge duplicates
df_cleaned = df.groupby(['fullVisitorId', 'visitId'], as_index=False).agg(agg_rules)

# Convert both to string and concatenate with a dash
df_cleaned['session_id'] = df_cleaned['fullVisitorId'].astype(str) + '-' + df_cleaned['visitId'].astype(str)

# Reorder columns: session_id first
cols = ['session_id'] + [col for col in df_cleaned.columns if col != 'session_id']
df_cleaned = df_cleaned[cols]

df_cleaned
```

```python
# Count duplicate combinations
duplicate_combos = df_cleaned.duplicated(subset=['fullVisitorId', 'visitId']).sum()
print(f"‚ôªÔ∏è Duplicate (fullVisitorId, visitId) pairs: {duplicate_combos}")
```
